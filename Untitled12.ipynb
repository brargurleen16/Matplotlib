```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('Teen_Mental_Health.csv')

# Show first rows
print(df.head())
```

```python
stress_counts = df['stress_level'].value_counts().sort_index()
plt.figure(figsize=(10, 6))

plt.bar(stress_counts.index, stress_counts.values, color = c[:len(stress_counts)])
plt.title('Stress Level Distribution')
plt.xlabel('Stress Level')
plt.ylabel('Number of Students')
plt.xticks(stress_counts.index)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
gender_counts = df['gender'].value_counts()
c=["red","blue","yellow","pink"]
plt.figure()
gender_counts.plot(kind='bar', color=c[:len(gender_counts)])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sleep_counts = df['sleep_hours'].value_counts().sort_index()

plt.figure()
sleep_counts.plot(kind='bar')
plt.title('Sleep Hours Distribution')
plt.xlabel('Sleep Hours')
plt.ylabel('Count')
plt.show()
```

```python
social_counts = df['daily_social_media_hours'].value_counts().sort_index()

plt.figure()
social_counts.plot(kind='bar')
plt.title('Social Media Usage (Hours)')
plt.xlabel('Hours')
plt.ylabel('Number of Students')
plt.show()
```

```python
anxiety_counts = df['anxiety_level'].value_counts()

plt.figure()
anxiety_counts.plot(kind='bar')
plt.title('Anxiety Level Distribution')
plt.xlabel('Anxiety Level')
plt.ylabel('Count')
plt.show()
```

```python
gender_counts = df['gender'].value_counts()

plt.figure()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
# 2. Stress Level Pie Chart
df['stress_level'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Stress Level Distribution')
plt.ylabel('')
plt.show()
```

```python
# 3. Anxiety Level Pie Chart
df['anxiety_level'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Anxiety Level Distribution')
plt.ylabel('')
plt.show()
```

```python
# 1. Sleep vs Stress
data = df.groupby('sleep_hours')['stress_level'].mean()
plt.plot(data.index, data.values, marker='o',color='purple')
plt.title('Sleep vs Stress')
plt.xlabel('Sleep Hours')
plt.ylabel('Avg Stress')
plt.grid()
plt.show()
```

```python
# 2. Sleep vs Anxiety
data = df.groupby('sleep_hours')['anxiety_level'].mean()
plt.plot(data.index, data.values, marker='o',color='pink')
plt.title('Sleep vs Anxiety')
plt.xlabel('Sleep Hours')
plt.ylabel('Avg Anxiety')
plt.grid()
plt.show()
```

```python
# 3. Social Media vs Stress
data = df.groupby('daily_social_media_hours')['stress_level'].mean()
plt.plot(data.index, data.values, marker='o')
plt.title('Social Media vs Stress')
plt.xlabel('Hours')
plt.ylabel('Avg Stress')
plt.grid()
plt.show()
```

```python
# 4. Social Media vs Anxiety
data = df.groupby('daily_social_media_hours')['anxiety_level'].mean()
plt.plot(data.index, data.values, marker='o')
plt.title('Social Media vs Anxiety')
plt.xlabel('Hours')
plt.ylabel('Avg Anxiety')
plt.grid()
plt.show()
```

```python
# 5. Stress vs Anxiety Trend
data = df.groupby('stress_level')['anxiety_level'].mean()
plt.plot(data.index, data.values, marker='o')
plt.title('Stress vs Anxiety')
plt.xlabel('Stress Level')
plt.ylabel('Avg Anxiety')
plt.grid()
plt.show()
```
